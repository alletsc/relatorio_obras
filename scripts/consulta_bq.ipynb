```python
import pandas as pd
import basedosdados as bd
```

```python
querie = """ SELECT
  *
FROM
  `rj-smi.infraestrutura_siscob_obras.obra` AS o
WHERE
  EXTRACT(
    YEAR
    FROM
      o.data_inicio
  ) >= 2021
  OR o.situacao IN ("EXECUTANDO", "SUSPENSA");"""
```

```python
df = bd.read_sql(querie, billing_project_id= "rj-segovi")
df.head()
```

```python
# salvar como csv
df.to_csv("obras.csv", index=False)
```
